```python
import matplotlib.pyplot as plt
```

```python
x=['1st','2nd','3rd','4th']
y=[25,50,75,100]
plt.bar(x,y,color='b')                 # color=red-r,green-g,blue-b
plt.title('Percentage Scheme as in Courses')
plt.xlabel('year')
plt.ylabel('Percentage')
plt.yticks(range(40,120,20))
plt.show()
```

```python
plt.plot(x,y,label='mpv',color='r')
y2=[88,87,83,85,90,85]
plt.plot(x,y2,label='sk',color='g')
plt.title('Academic performance ')
plt.xlabel('Semester')
plt.ylabel('Percentage')
plt.legend()
plt.show()
```

```python
age=[10,20,25,30,35,39,11,8,9,8,16,13,46,67,34,55,45,54,56,65,44,43,36,22,11,34,66,88,23,77,99]
group_age=[0,20,40,60,80,100]
plt.hist(age,group_age,rwidth=0.9,color='r')
plt.xticks(group_age)            # x-axis user defind range
plt.yticks(range(0,12,2))       # here u can user defind range of y-axis 
plt.ylabel('No.of person ')
plt.xlabel('age range')
plt.title('Age group information ')

plt.show()
```

```python
h=[62,72,77,74,69]
w=[155,220,240,195,175]
plt.scatter(w,h,color='r')
plt.xlabel('height')
plt.ylabel('weight')
plt.show()
```

```python
activities=['sleeping','coding','entertainment','social_media']
slices=[8,12,5,4]
plt.pie(slices,labels=activities,autopct='%1.2f%%',shadow=True)   # here autopct used for showing percentage ratio
plt.title(' Daily routine: mpv@coder')                                       # shadow used for 3d view of result 
plt.show()
```
